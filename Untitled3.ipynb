```python
def repl_99999(col):

    real_values = col.loc[(~col.isna()) & (col != 99999.00000)]

    true_median = real_values.median()

    return col.replace(99999.00000, true_median)
```

```python
def scale_logreg_cv(X, y):
    
    # create train and holdout set
    X_train, X_holdout, y_train, y_holdout = train_test_split(X, y, test_size=0.7, random_state=727)
    
    scaler = StandardScaler()
    X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), 
                            index=X_train.index, 
                            columns=X_train.columns)
    
    logreg = LogisticRegression(max_iter=100000)
    
    return cross_validate(logreg, X_train_scaled, y_train, scoring=['accuracy', 'precision', 'recall', 'f1'])
```
