```python
def repl_99999(col):

    real_values = col.loc[(~col.isna()) & (col != 99999.00000)]

    true_median = real_values.median()

    return col.replace(99999.00000, true_median)
```

```python
def replace_with_median(col, value_to_replace):
    
    real_values = col.loc[(~col.isna()) & (col != value_to_replace)]

    true_median = real_values.median()

    return col.replace(value_to_replace, true_median)
```

```python
# create holdout set

X_training, X_holdout, y_training, y_holdout = train_test_split(X, y, test_size=0.7, random_state=727)
```

```python
# try a logistic regression
    
X_train, X_test, y_train, y_test = train_test_split(X_training, y_training, test_size=0.3)

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)
    
logreg = LogisticRegression(max_iter=10000)

logreg.fit(X_train, y_train)

y_pred_train = logreg.predict(X_train)
y_pred_test = logreg.predict(X_test)

print(recall_score(y_train, y_pred_train, pos_label=1))
print(recall_score(y_test, y_pred_test, pos_label=1))
    
plot_confusion_matrix(logreg, X_train, y_train)
plt.title('Train set')
    
plot_confusion_matrix(logreg, X_test, y_test)
plt.title('Test set')

cross_validate(logreg, X, y, scoring=['accuracy', 'precision', 'recall', 'f1'], return_train_score=True)
```

```python
# try a complement naive Bayes

X_train, X_test, y_train, y_test = train_test_split(X_training, y_training, test_size=0.3)
    
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)
    
CompNB = ComplementNB()
    
CompNB.fit(X_train, y_train)

y_pred_train = CompNB.predict(X_train)
y_pred_test = CompNB.predict(X_test)

print(recall_score(y_train, y_pred_train, pos_label=1))
print(recall_score(y_test, y_pred_test, pos_label=1))
    
plot_confusion_matrix(CompNB, X_train, y_train)
plt.title('Train set')
    
plot_confusion_matrix(CompNB, X_test, y_test)
plt.title('Test set')

cross_validate(CompNB, X, y, scoring=['accuracy', 'precision', 'recall', 'f1'], return_train_score=True)
```
